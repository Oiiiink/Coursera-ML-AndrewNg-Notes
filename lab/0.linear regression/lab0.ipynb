### An easy try on Linear Regression

```python
# import library and load datas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r'D:\GitHub\ML\Coursera-ML-AndrewNg-Notes\lab\0.linear regression\test.csv')
df.head()
```

```python
def cost_function(x, y, w, b):
    predictions = w*x + b
    return np.sum((predictions-y)**2) / 2*len(x)
def derivative_w_and_b(x, y, w, b):
    """ calculate the derivative of J(w) and J(b)
    parameter:
        x: features
        y: targets
        w: former w
        b: former b
    
    return:
        w_der: the derivative of J(w)
        b_der: the derivative of J(b)
    """
    assert len(x)==len(y)
    m = len(x)
    predictions = w * x + b
    errors = predictions - y
    w_der = (1/m) * np.sum(errors*x)
    b_der = (1/m) * np.sum(errors)
    return w_der, b_der
def linear_regression(x, y, alpha=0.7, iterations=10000):
    costs = []
    times = []
    i = 0
    w, b = 0, 0
    while i < iterations:
        i += 1
        w0, b0 = w, b
        w_der, b_der = derivative_w_and_b(x, y, w, b)
        w, b = w-alpha*w_der, b-alpha*b_der
        if w0==w and b0==b:
            # reach local minimum
            print(cost_function(x, y, w, b))
            break
    return w, b
```

```python
training_set = df['x']
targets = df['y']
plt.plot(training_set, targets, color='r')
plt.xlabel('x')
plt.ylabel('y')
w, b= linear_regression(np.array(training_set), np.array(targets), 0.0005)
plt.plot(training_set, w*training_set+b, color='b')
plt.show()
print('the final function is y = ', str(w), '*x ' + str(b))
```
