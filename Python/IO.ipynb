**使用Python来操作文件**


```python
file_sample = open(file = 'ex.txt', mode = 'r', encoding = 'utf-8') # 当文件中有中文时，需要第三个参数

print(file_sample.read(3))
print(file_sample.readline())
context = file_sample.readlines()
for c in context:
    print(c)

file_sample.close() # 必须！！！

# 效果同上但是会自动关闭
with open(file = 'ex.txt', mode = 'r', encoding = 'utf-8') as file_sample:
    a = 10

```
